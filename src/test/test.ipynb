```python
from src.data.preprocessing import create_preprocessing_pipeline
import pandas as pd

df = pd.read_csv("../datasets/bronze/Students_Social_Media_Addiction.csv")
```

```python

X_transformed = pipeline.fit_transform(df.head(10))
col_transformer = pipeline.named_steps["preprocessing"]
column_names = col_transformer.get_feature_names_out()
print(column_names)
print(column_names.shape)
X_df = pd.DataFrame(X_transformed, columns=column_names)
X_df.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

preprocessing_pipeline = create_preprocessing_pipeline()

# Random Forest pipeline
rfr_pipeline = Pipeline([
    ("preprocessing", preprocessing_pipeline),  # jouw bestaande preprocessing pipeline
    ("model", RandomForestRegressor(random_state=42))
])

# Linear Regression pipeline
lr_pipeline = Pipeline([
    ("preprocessing", preprocessing_pipeline),
    ("model", LinearRegression())
])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("Addicted_Score", axis=1)
y = df["Addicted_Score"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30)
```

```python
rfr_pipeline.fit(X_train, y_train)
lr_pipeline.fit(X_train, y_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


y_pred_rfr = rfr_pipeline.predict(X_test)
y_pred_lr = lr_pipeline.predict(X_test)

print("Random Forest RMSE:", mean_squared_error(y_test, y_pred_rfr))
print("Random Forest R2:", r2_score(y_test, y_pred_rfr))

print("Linear Regression RMSE:", mean_squared_error(y_test, y_pred_lr))
print("Linear Regression R2:", r2_score(y_test, y_pred_lr))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.linear_model import Ridge


```

```python
rfr_pipeline = Pipeline([
    ("preprocessing", preprocessing_pipeline),  # jouw preprocessing pipeline
    ("model", RandomForestRegressor(random_state=42))
])

rr_pipeline = Pipeline([
    ("preprocessing", preprocessing_pipeline),
    ("model", Ridge())
])
```

```python
# RFR
param_grid_rfr = {
    "model__n_estimators": [50, 100, 200],
    "model__max_depth": [None, 10, 20],
    "model__min_samples_split": [2, 5, 10]
}

grid_search_rfr = GridSearchCV(
    estimator=rfr_pipeline,
    param_grid=param_grid_rfr,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=2
)
```

```python
param_grid_rr = {
    "model__alpha": [0.1, 1.0, 10.0, 100.0]
}

grid_search_rr = GridSearchCV(
    estimator=rr_pipeline,
    param_grid=param_grid_lr,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=2
)
```

```python
# Stap 3: Fitten en evalueren
# Fit Random Forest GridSearch
grid_search_rfr.fit(X_train, y_train)
print("Best RF params:", grid_search_rfr.best_params_)
print("Best RF RMSE (cv):", -grid_search_rfr.best_score_)

# Predict en eval op testset
y_pred_rfr = grid_search_rfr.best_estimator_.predict(X_test)
print("Test RF RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rfr)))
print("Test RF R2:", r2_score(y_test, y_pred_rfr))


# Fit Linear Regression (Ridge) GridSearch
grid_search_rr.fit(X_train, y_train)
print("Best LR params:", grid_search_rr.best_params_)
print("Best LR RMSE (cv):", -grid_search_rr.best_score_)

# Predict en eval op testset
y_pred_lr = grid_search_rr.best_estimator_.predict(X_test)
print("Test LR RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print("Test LR R2:", r2_score(y_test, y_pred_lr))

```

```python
import mlflow
import mlflow.sklearn
from sklearn.metrics import mean_squared_error

experiment_name = "socialmedia_addiction_models"
mlflow.set_experiment(experiment_name)

with mlflow.start_run(run_name="RandomForest") as run:
    grid_search_rfr.fit(X_train, y_train)
    y_pred = grid_search_rfr.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mlflow.log_metric("mse", mse)
    
    r2 = r2_score(y_test, y_pred)
    mlflow.log_metric("r2", r2)

    # Log parameters (optioneel)
    mlflow.log_param("model_type", "RandomForest")
    mlflow.log_param("random_state", 42)
    
    # Log het volledige pipeline model (preprocessing + regressor)
    mlflow.sklearn.log_model(grid_search_rfr, "model")
    mlflow.log_params(grid_search_rfr.best_params_)
    
    # Tag het run (extra metadata)
    mlflow.set_tag("model_type", "RandomForest")
    mlflow.set_tag("flow_id", flow_id)
```

```python
with mlflow.start_run(run_name="Ridge") as run:
    grid_search_rr.fit(X_train, y_train)
    y_pred = grid_search_rr.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mlflow.log_metric("mse", mse)
    
    r2 = r2_score(y_test, y_pred)
    mlflow.log_metric("r2", r2)
    
    mlflow.log_param("model_type", "Ridge")
    mlflow.log_params(grid_search_rr.best_params_)
    
    mlflow.sklearn.log_model(grid_search_rr, "model")
    
    mlflow.set_tag("model_type", "Ridge")
    mlflow.set_tag("flow_id", flow_id)
```

```python
def evaluate_latest_flow(flow_id, experiment_name="your_experiment", metric="rmse"):
    client = mlflow.tracking.MlflowClient()
    experiment = client.get_experiment_by_name(experiment_name)
    
    runs = client.search_runs(experiment_ids=[experiment.experiment_id], filter_string=f'tags.flow_id = "{flow_id}"')
    best_run = min(runs, key=lambda r: r.data.metrics.get(metric, float('inf')))

    return best_run.info.run_id, best_run.data.tags.get("model_type")
```

```python
best_model = evaluate_latest_flow(flow_id, experiment_name, "mse")
best_model
```
